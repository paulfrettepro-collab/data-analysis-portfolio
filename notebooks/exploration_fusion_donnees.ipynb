```python
import pandas as pd

def assess_email_validity(email, firstname, lastname):
    probability = 1.0

    disposable_domains = ['mailinator.com', 'guerrillamail.com', 'yopmail.com']
    if any(domain in email for domain in disposable_domains):
        probability -= 0.4

    if 'missing_email' in email:
        probability = 0.0

    if len(email) > 100 or any(c in email for c in ['%', '$', '#', '@', '&']):
        probability -= 0.2

    if pd.notna(firstname) and pd.notna(lastname):
        if any(domain in email for domain in disposable_domains):
            if isinstance(firstname, str) and isinstance(lastname, str):
              if len(firstname) > 20 or len(lastname) > 20:
                  probability -= 0.3
    return max(0.0, min(1.0, probability))

df['email_validity_probability'] = df.apply(lambda row: assess_email_validity(row['email_cleaned'], row['firstname'], row['lastname']), axis=1)

display(df.head())
```
